```python
import cv2
from matplotlib import pyplot as plt
# Load an image (make sure you have an image file in the same folder)
img = cv2.imread('butterfly.jpg',1)  # replace with your image name
# Convert BGR (OpenCV default) to RGB for Matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display
plt.imshow(img_rgb)
#plt.imshow(img)
plt.title('Loaded Image with OpenCV')
plt.axis('on')
plt.show()
```

```python
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Resize
resized = cv2.resize(img, (300, 200))
plt.imshow(resized)
#plt.imshow(img)
plt.title('Loaded Image with OpenCV')
plt.axis('off')
plt.show()
print(resized.shape)
```

```python
# Rotate (90 clockwise)
rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.imshow(rotated)
#plt.imshow(img)
plt.title('Loaded Image with OpenCV')
plt.axis('off')
plt.show()
```

```python
# Flip (0: vertical, 1: horizontal)
flipped = cv2.flip(img, 0)
plt.imshow(flipped)
#plt.imshow(img)
plt.title('Loaded Image with OpenCV')
plt.axis('off')
plt.show()
```

```python
blur = cv2.GaussianBlur(img, (7, 7), 0)
plt.imshow(blur)
#plt.imshow(img)
plt.title('Image with Blur')
plt.axis('off')
plt.show()
```
